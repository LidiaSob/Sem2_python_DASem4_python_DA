# Условие 3:
# 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
sns.scatterplot(data=houses, x='condition', y='price', hue='price');
```

Визуальная оценка графика позволяется сделать вывод, что существует корреляция между состоянием дома и его ценой. В данных наблюдаются выбросы, а зависимость - нелинейная. Для более корректной оценки нужно учесть больше переменных.

Видно, что на отрезке оси х от 1 до 2, чем хуже состояние дома, тем он дешевле, что логично, однако для оценки отрезка 3-5 нужно больше данных.

```python
sns.scatterplot(data=houses, y='sqft_living', x='price', hue='price');
```

На данном графике также наблюдаются выбросовые значения. В отличие от графика зависимости цены от состояния дома, на этом графике зависимость линейная (визуально, все равно нужна тщательная оценка).

Таким образом, можно сказать, что чем больше площадь дома, тем выше цена (что логично).

```python
sns.scatterplot(data=houses, x='yr_built', y='price', hue='price');
```

Визуальная оценка данного графика, по моему мнению, абсолютно бесполезна, нужно добавить больше переменных, влияющих на цену.

```python
sns.scatterplot(data=houses, x='view', y='price', hue='price', legend=False);
```

В целом, распеределение вид-цена - равномерное, в данном случае видно, что такой параметр, как "вид" не является решающий при ценообразовании, поэтому, визуально оценивать зависимость нет смысла.

```python
sns.scatterplot(data=houses, x='floors', y='price', hue='price', legend=False);
```

На данном графике можно заметить, что зависимость нелинейна. Дома с большим количеством этажей стоят дешевле, чем с меньшим. В данном случае необходимо учитывать больше факторов. Возможно они пользуются меньшим спросом, их состояние или год постройки неприемлемы для покупателей.
Дороже всего стоят дома с 2 этажами, возможно они пользуются наибольшим спросом, но это уже дополнительная переменная.
